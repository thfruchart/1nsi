# Question 1

Ecrire une fonction rendu_monnaie qui :  
* prend en arguments : 
  * la somme à rendre `somme`
  * la liste des valeurs des pièces `systeme` 
* et qui renvoie la liste des effectifs des pièces à rendre, pour chacune des valeurs du système, avec un algorithme glouton.

Par exemple :  
```
>>> rendu_monnaie(93, [50,20,10,5,2,1])  
[1,2,0,0,1,1]
```

````
def rendu_monnaie( somme , systeme  ) : 
    monnaie_list = [ 0 for i in range(len(systeme))]
    reste_a_rendre = somme
    i = 0
    while reste_a_rendre > 0 :
        piece = systeme[i]
        nb = reste_a_rendre // piece
        reste_a_rendre = reste_a_rendre % piece
        print('Rendre', nb, 'fois la valeur', piece)
        monnaie_list[i] = nb
        i += 1
    
    return monnaie_list
````

````
def rendu_monnaie( somme , systeme  ) : 
    monnaie_list = [ 0 for i in range(len(systeme))]
    i = 0
    while somme > 0 :
        piece = systeme[i]
        monnaie_list[i] = somme // piece
        somme %=  piece
        i += 1
    return monnaie_list
````

### Ecrire une nouvelle version de la fonction rendu_monnaie, qui :
* prend en arguments : 
  * la somme à rendre `somme`
  * la liste des valeurs des pièces `systeme` 
* et qui renvoie la liste des pièces à rendre, la plus courte possible.

Par exemple :  
```
>>> rendu_monnaie(93, [50,20,10,5,2,1])  
[50,20,20,2,1]
```



````
def rendu_monnaie( somme , systeme  ) : 
    monnaie_list = []
    i = 0
    while somme > 0 :
        piece = systeme[i]
        nb = somme // piece
        for _ in range(nb):
            monnaie_list.append(piece)
        somme %=  piece
        i += 1
    return monnaie_list
````

````
rendu_monnaie(93, systeme_canonique)
````

```
import csv
import matplotlib.pyplot as plt
from operator import itemgetter


def creerListes(fichier):
    #fichier est une chaine de caractères
    #assez clair et joli : on recup des listes Python faciles à traiter
        
    temp= open(fichier,newline='')
    L=list(csv.reader(temp))
                       
    liste_sepal=[float(L[i][2]) for i in range(1,len(L))]
    liste_petal=[float(L[i][4]) for i in range(1,len(L))]
    liste_variete=[L[i][5] for i in range(1,len(L))]
            
    return liste_sepal,liste_petal,liste_variete

def affichePoints():
    x,y,variete=creerListes("iris.csv")
    plt.plot(x[0:50],y[0:50],'ro',label='Setosa')
    plt.plot(x[50:100],y[50:100],'ob',label='Versicolor')
    plt.plot(x[100:],y[100:],'og',label='Virginica')
    plt.title("Fleurs d'iris")
    plt.grid()
    plt.xlabel('Largeur des sépales (cm)')
    plt.ylabel('Largeur des pétales (cm)')
    plt.legend()
    plt.show()


# définition des listes d'abscisses, ordonnées, et valeurs
x,y,v=creerListes("iris.csv")

affichePoints()
```

#### distance et calcule_distances

```
def distance(x1,y1,x2,y2):
    """cette fonction prend en paramètres :
    x1,y1,x2,y2 de type int ou float 
    
    elle renvoie la distance entre les points P1 (x1,y1)  et P2(x2,y2) 
    """
    return ((x2-x1)**2+(y2-y1)**2)**0.5


def calcule_distances(x,y,liste_x,liste_y,liste_valeurs):
    """ cette fonction prend en paramètres : 
    x,y de type int ou float
    liste_x,liste_y,liste_valeurs : trois listes de même longueur.
    
    elle renvoie une liste de listes
    contenant : un numéro entier, la distance entre P(x,y) et Pi(xi,yi), la classe n°i
    """
    tab=[]
    for i in range(len(liste_valeurs)):
        tab.append([i+1, distance(x,y,liste_x[i],liste_y[i]), liste_valeurs[i]])
    return tab
```

```
# Tester un exemple ci-dessous en complétant les trois listes

calcule_distances(1,2, [0,1,2,1], [0,1,4,2], ['Setosa','Setosa','Versicolor','Versicolor'])
```

#### tri et classe

```
def tri(tab):
    return sorted(tab,key=itemgetter(1))

def classe(k,tableauTrie):
    """ la fonction prend en paramètres : 
    k de type int
    tableauTrie de type liste de listes
    
    elle renvoie un dictionnaire
    contenant le nombre de fleurs de chaque type seulement pour les k PREMIERES VALEURS DU TABLEAU  
    """
    tab=tableauTrie[:k]
    dico={'Setosa':0,'Versicolor':0,'Virginica':0}
    for elm in tab:
        dico[elm[2]]+=1
    return dico
```

```
tablo = [(10,20),(40,30),(70,10), (50,15)]
tableau = tri(tablo)
print(tableau)
```

```
# Tester un exemple en complétant la liste
tablo = [[1, 2.23606797749979, 'Setosa'], 
         [2, 1.0, 'Setosa'], 
         [3, 2.23606797749979, 'Versicolor'], 
         [4, 0.0, 'Versicolor']]
tableau = tri(tablo)
print(tableau)
d = classe(3,tableau)
print(d)
```

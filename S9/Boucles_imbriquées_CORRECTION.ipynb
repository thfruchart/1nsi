Dans chacun des programmes suivants, déterminer en fonction de n quelle sera la valeur affichée.

```
# programme 1
n = int(input('Entrer un entier positif n : '))

cpt = 0
for i in range(n):
    cpt += 10

print(cpt)
print(10*n)
# conclusion : le compteur est égal à 10*n
```

```
# programme 2
n = int(input('Entrer un entier positif n : '))

cpt = 0
for i in range(n):
    cpt += 10
for j in range(n):
    cpt += 100
print(cpt, 110*n)
```

```
# programme 3
n = int(input('Entrer un entier positif n : '))

cpt = 0
for i in range(n):
    
    for j in range(n):
        cpt += 1
print(cpt, n**2)
```

```
# programme 4
n = int(input('Entrer un entier positif n : '))

cpt = 0
for i in range(n):
    cpt += 2
    for j in range(n):
        cpt += 1
print(cpt,0+2*n+1*n*n)
# conclusion : le compteur est égal à n²+2n
```

```
# programme 5
n = int(input('Entrer un entier positif n : '))

cpt = 0
for i in range(n):
    for j in range(i): # i à la place de n
        cpt += 1
print(cpt, (n-1)*n//2) 
```

```
# programme 5 bis
n = int(input('Entrer un entier positif n : '))

cpt = 0
for i in range(n):
    for j in range(i): # i à la place de n
        cpt += 1
        print('*', end=' ')
    print('.')
print(cpt, (n-1)*n//2) 
```

```
# programme 6 (DIFFICILE)
n = int(input('Entrer un entier positif n : '))

cpt = 0
for i in range(n):
    cpt += 3
    for j in range(i): # i à la place de n
        cpt += 2
        print('*', end=' ')
    print('.')
print(cpt, 0+2*n+1*n*n)
```

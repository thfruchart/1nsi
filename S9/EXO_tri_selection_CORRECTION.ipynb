# Exercice :  tri par sélection

La fonction suivante implémente l'algorithme du tri par sélection

```python
def tri_par_selection(t):
    # t[0:i] est trié et ses valeurs sont inférieurs ou égales à celles de t[i:]
    for i in range(len(t)-1): 
        # on cherche l'indice du minimum de t[i:]
        i_min = i 
        for j in range(i+1, len(t)): 
            if t[j] < t[i_min] :
                i_min = j
        if i_min != i :
            # on range le minimum à l'indice i (avant de poursuivre à l'indice i+1)
            t[i], t[i_min] = t[i_min], t[i] 
    return t
```

On donne le tableau suivant, dans lequel chaque item est un tuple

```python
tab = [ ('Lucie', 'Agra', '01/03/2006'), ('Richard', 'Bert', '21/02/2006'), ('Lara', 'Croft', '29/07/2005'), ('Claude','Imbre','31/12/2005')]
```

### Généralisation (1) : croissant ou décroissant

On donne un tableau `t1` et une fonction `tri` qui admet deux  paramètres : 
* un tableau `t`
* un booléen `croissant` (valeur par défaut : `True`)

```python
t1 = [10,50,30,80,60,70,20]

def tri(t, croissant = True ):
    for i in range(len(t)-1): 
        # on cherche l'indice d'un extremum de t[i:]
        i_extreme = i 
        for j in range(i+1, len(t)): 
            if (t[j] < t[i_extreme] ) == croissant :
                i_extreme = j
        if i_extreme != i :
            # on range l'extremum à l'indice i (avant de poursuivre à l'indice i+1)
            t[i], t[i_extreme] = t[i_extreme], t[i] 
    return t
```

```python
tri(t1, True)
```

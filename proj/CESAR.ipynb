# Chiffrement de César
* On souhaite chiffrer un message pour le communiquer de manière confidentielle
* Le principe du chiffrement de César consiste à décaler toutes les lettres de l'alphabet d'un certain nombre, que l'on notera d (pour décalage)
* chaque lettre du message d'origine est remplacée par la lettre décalée de "d" places dans l'alphabet.

```
ABCDEFGHIJKLMNOPQRSTUVWXYZ
```

Exemple1 : chiffrer le message "HELLO" avec d = 3

Exemple2 : déchiffrer le message RXL avec d = 3

Exemple3 : chiffrer le mot "YES" avec d = 4

### réalisation du chiffrement avec python

* plusieurs méthodes sont possibles
* pour chiffrer facilement un message on peut utiliser un dictionnaire de chiffrement dans lequel : 
  * les clés sont les lettres "en clair"
  * les valeurs sont les lettres "chiffrées"
  * on peut alors chiffrer un message lettre après lettre
* on peut procéder de même pour le déchiffrement !

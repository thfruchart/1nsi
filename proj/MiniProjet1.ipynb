## Mini-projet

#### Conversion décimale => binaire pour les nombres flottants.  
En Python, il est facile de déterminer l'écriture binaire d'un nombre entier, avec la fonction `bin`.
Tester les exemples ci-dessous :

```python
bin(25)
```

```python
bin(2**10)
```

```python
bin(0.25)
```

=> Problème : python ne gère pas la conversion binaire des nombres flottants

### Objectif de ce mini projet :
#### écrire une fonction qui affiche à l'écran l'écriture binaire d'un nombre x.

#### Remarques importantes

* Puisque certains nombres flottants n'ont pas d'écriture binaire finie, il faudra limiter le nombre de chiffres affichés.
* on pourra commencer par supposer que  `x` est compris entre 0 et 1.
* on pourra ensuite améliorer la fonction en traitant les questions suivantes :
  1. écrire une fonction `conversion_binaire(x:float)` qui **renvoye** l'écriture binaire au format `str` au lieu de l'afficher à l'écran.
  2. laisser l'utilisateur choisir le nombre de chiffres après la virgule qu'il souhaite obtenir
  3. traiter le cas où x est supérieur à 1

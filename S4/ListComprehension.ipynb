## listes définies par compréhension

```python
t1 = [3, 5, -5, 0, -4, -3, 1, 0]

t2 = [ v**2 for v in t1]
```

```python
t2
```

```python
t1 = [3, 5, -5, 0, -4, -3, 1, 0]

t3 = [ v for v in t1 if v <= 0]
```

```python
t3
```

```python
tab1 = [(7, 2), (5, 4), (1, 2), (1, 8), (0, 9)]

tab2 = [ x[0]+x[1] for x in tab1]
```

```python
tab2
```

```python
tab1 = [(7, 2), (5, 4), (1, 2), (1, 8), (0, 9)]

tab3 = [ v for v in tab1 if v[0]>v[1]]
```

```python
tab3
```

```python
tab1 = [(7, 2), (5, 4), (1, 2), (1, 8), (0, 9)]

tab4 = [ a for a,b in tab1]
```

# TP prénoms

Un [fichier csv](https://raw.githubusercontent.com/thfruchart/1nsi/main/S4/nat2021.csv) contient, pour chaque année et pour chaque prénom recensé, le nombre de fois où ce prénom a été déclaré à l'état civil français.

Si un prénom a été donné moins de deux fois en une année, il n'est pas recensé, mais il est comptabilisé dans les "prénoms rares" de l'année.

Certaines années sont notées XXXX : cette notation regroupe toutes les années où un prénom a été considéré comme "prénom rare", et donne le nombre total sur l'ensemble de ces années.

### lecture du fichier csv
Le code ci-dessous permet d'importer le contenu du fichier dans une liste de dictionnaires.

```python
import csv

def lecture_fichier(nom_fichier):
    "lit un fichier csv et renvoie une liste de dictionnaires dont les clés sont les descripteurs"
    with open(nom_fichier, "r") as fichier:
        return [ligne for ligne in csv.DictReader(fichier, delimiter=',')]
```

```python
prenoms = lecture_fichier('nat2021.csv')
```

# Consultation par année

Dans cette partie, l'utilisateur doit pouvoir :
* choisir une année
* rechercher quels sont les prénoms les plus fréquents pour cette année
* il doit pouvoir effectuer cette recherche
   * soit avec les prénoms masculins
   * soit avec les prénoms féminins
   * soit avec tous les prénoms 
* il doit également pouvoir choisir le nombre de prénoms à afficher (par défaut : 10)

```python
def critere_nombre(ligne):
    return int(ligne['nombre'])

def recherche_par_annee(annee, sexe = None, nb = 10) :
    # sélection des lignes de la table
    selection = [  ligne  for ligne in prenoms if ligne['année'] == str(annee) ]
    
    # selection par sexe
    if sexe == 1 or sexe == 2 :
        selection =  [  ligne  for ligne in selection if ligne['sexe'] == str(sexe) ]
        
    # tri des valeurs
    resultat = sorted(selection, key = critere_nombre, reverse = True)
    
    return resultat[0:nb]
```
